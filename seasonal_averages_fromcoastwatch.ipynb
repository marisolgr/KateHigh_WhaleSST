```python
# Modules
import warnings
warnings.simplefilter('ignore') # filter some warning messages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
```

```python
nm = ['winter','spring','summer','fall']
```

```python
for i in nm:
    ds = xr.open_mfdataset('~/Downloads/'+i+'/*.nc')
    ds2 = ds.sel(lat=slice(40,37),lon=slice(-124.5, -121)).mean('time')
    ds2.to_netcdf(i+'_SST_1995-2019.nc')
    ds2.sea_surface_temperature.plot()
    plt.show()
```
